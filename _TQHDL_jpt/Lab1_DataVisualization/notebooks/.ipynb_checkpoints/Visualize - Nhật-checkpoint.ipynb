```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_csv('../data/Processed.csv')
data_df.sample(n=5)
```

## Question: Tổng chi tiêu của sinh viên từng chuyên ngành theo năm học thay đổi như thế nào?

### 1. Purpose
One of the objectives of analyzing the total expenditure of students by major and academic year is to gain a deeper understanding of how various factors influence students' spending during their academic journey. Some specific objectives may include:
- Understanding spending trends: Analyzing total expenditure by academic year can help us grasp an overall trend in how students' spending changes over time. This may shed light on students' development throughout their academic years and identify factors that may influence their spending levels.
- Identifying differences between majors: Comparing total expenditure across different majors can help us understand the differences in spending levels among student groups. This could reflect factors such as average post-graduation income, tuition costs, or students' spending preferences based on their majors.

### 2. Analyze

```python
data_df['Total Spending'] = data_df.iloc[:, 5:-1].sum(axis=1)
grouped = data_df.groupby(['Major', 'Year In School']).mean(numeric_only=True)
grouped
```

### 3. Visualize

```python
# Plotting
plt.figure(figsize=(10, 6))

# Iterate through each major
for major, group in grouped.groupby(level=0):
    plt.plot(group.index.get_level_values('Year In School'), group['Total Spending'], label=major)

plt.title('Total Spending by Year in School for Each Major')
plt.xlabel('Year In School')
plt.ylabel('Total Spending')
plt.legend(title='Major', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 4. Observation


- Psychology seems to be stable: The stability in spending of Psychology students may reflect a consistent level of expenditure over the academic years.  
- Economics shows the largest fluctuations: The significant fluctuations in spending among Economics students may indicate variations in their spending habits, especially in the second and third years. There could be both internal and external factors causing these fluctuations, such as changes in income levels, tuition fees, or shifting spending priorities.
- Computer Science consistently is one of major spends the most: The tendency for Computer Science students to have the highest spending may reflect the specific costs associated with technology, hardware, and software required for studies and research in this field.  
- Economics consistently is one of major spends the least: The low variability in spending among Economics students may reflect caution or emphasis on financial management and savings.
