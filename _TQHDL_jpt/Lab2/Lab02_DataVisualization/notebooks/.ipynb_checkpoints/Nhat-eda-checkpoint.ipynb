## Câu hỏi: Sự thay đổi của tổng số lượng khách hàng theo từng năm như thế nào?

### 1. Mục đích

   - Phân tích xu hướng tăng trưởng hoặc suy giảm của số lượng hành khách đi máy bay từ 2000-2020.  
   - Đánh giá tác động của các sự kiện lớn (như khủng hoảng tài chính, khủng bố 11/9, COVID-19) đến ngành hàng không.   
   - Đánh giá khả năng phục hồi của ngành hàng không sau các giai đoạn suy giảm.  
   - Sử dụng dữ liệu lịch sử để dự báo xu hướng tương lai và hỗ trợ lập kế hoạch chiến lược. 

### 2. Phân tích

```python
passengers_per_year = df.groupby('YEAR')['Sum_PASSENGERS'].sum().reset_index()

# Display the results
print(passengers_per_year)
```

### 3. Trực quan hóa 

Lựa chọn biểu đồ đường để trực quan, vì:
- Biểu đồ đường thể hiện nhanh được xu hướng tăng hay giảm qua các năm

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.plot(passengers_per_year['YEAR'], passengers_per_year['Sum_PASSENGERS'], marker='o')
plt.title('Total Number of Air Passengers per Year (2000-2020)')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.grid(True)

# Set the x-ticks to display every year
plt.xticks(range(2000, 2021, 1))  # Sets the x-axis ticks to show every year from 2000 to 2020

plt.show()
```

### 4. Kết luận


1. **Xu Hướng Tăng Trưởng Tổng Thể**:
   - Từ năm 2000 đến 2019, có sự tăng trưởng rõ rệt về số lượng hành khách đi máy bay. Tổng số hành khách tăng từ khoảng 741 triệu vào năm 2000 lên hơn 1 tỷ vào năm 2018 và 1.05 tỷ vào năm 2019.

2. **Ảnh Hưởng Của Các Sự Kiện Kinh Tế và Toàn Cầu**:
   - **2001-2002**: Có sự giảm đáng kể số lượng hành khách từ 741 triệu vào năm 2000 xuống còn 688 triệu vào năm 2001 và tiếp tục giảm xuống 677 triệu vào năm 2002. Sự giảm này có thể do hậu quả của vụ tấn công khủng bố ngày 11 tháng 9 năm 2001, gây ảnh hưởng lớn đến ngành hàng không toàn cầu.
   - **2008-2009**: Một sự sụt giảm khác xảy ra trong cuộc khủng hoảng tài chính toàn cầu, với số lượng hành khách giảm từ 814 triệu vào năm 2008 xuống còn 771 triệu vào năm 2009.

3. **Phục Hồi và Tăng Trưởng**:
   - Sau các giai đoạn giảm vào năm 2002 và 2009, số lượng hành khách đã cho thấy xu hướng phục hồi. Số lượng hành khách tăng từ 710 triệu vào năm 2003 lên 774 triệu vào năm 2004, và từ 771 triệu vào năm 2009 lên 792 triệu vào năm 2010.
   - Từ năm 2010 trở đi, có sự tăng trưởng ổn định gần như mỗi năm, cho thấy sự phát triển mạnh mẽ của ngành hàng không.

4. **Các Năm Đỉnh Điểm**:
   - Số lượng hành khách cao nhất được ghi nhận vào năm 2019, với hơn 1.05 tỷ hành khách. Đây là đỉnh cao của xu hướng tăng trưởng trước khi có sự sụt giảm đáng kể vào năm 2020.

5. **Ảnh Hưởng Của Đại Dịch COVID-19**:
   - Năm 2020, có sự giảm mạnh số lượng hành khách xuống còn 338 triệu, thấp hơn rất nhiều so với các năm trước đó. Sự giảm đột ngột này là do đại dịch COVID-19, dẫn đến các hạn chế đi lại rộng rãi, phong tỏa và sự giảm sút nghiêm trọng trong việc đi lại bằng đường hàng không toàn cầu.
