```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/Processed.csv")
df.head()
```

## Câu hỏi 1: Tổng số hành khách (Sum_PASSENGERS) của 10 hãng hàng không (AIRLINE_ID) có tổng số hành khách tính từ 2000 đến 2020 là nhiều nhất qua từng năm thay đổi như thế nào

### 1. Mục đích
- Biết được sự thay đổi của số lượng hành khách của 10 hãng hàng không có nhiều khách nhất (tính từ 2000-2020)
- Qua đó có thể phân tích được tình hình chung của việc khai thác các chuyến bay của 10 hãng hàng không có nhiều hành khách nhất (tính từ 2000-2020)

### 2. Phân tích

```python
sum_passengers = df.groupby('AIRLINE_ID')['Sum_PASSENGERS'].sum()
top_airlines = sum_passengers.nlargest(10).index
filtered_df = df[df['AIRLINE_ID'].isin(top_airlines)]
pivot_df = filtered_df.pivot_table(index='YEAR', columns='AIRLINE_ID', values='Sum_PASSENGERS', aggfunc='sum')
```

### 3. Trực quan hóa
Lựa chọn biểu đồ đường để trực quan, vì:
- Biểu đồ đường thể hiện nhanh được xu hướng tăng hay giam qua các năm

```python
plt.figure(figsize=(14, 6))

for airline_id in top_airlines:
    plt.plot(pivot_df.index, pivot_df[airline_id], label=df[df['AIRLINE_ID'] == airline_id]['CARRIER_NAME'].iloc[0])

years = pivot_df.index.tolist()
plt.xticks(years, [str(year) for year in years])

plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers per Year for Top 10 Airlines')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. Kết luận
- Từ năm 2000-2019, số hành khách di chuyển bằng máy bay có xu hướng tăng
- Năm 2020 số lượng hành khách thực hiện các chuyến bay giảm đột ngột, đúng với thực tế vì dịch bệnh covid-19
- Năm 2009, 2011, 2015 lần lượt các hãng hàng không Northwest Airlines, Continental Air Lines, US Airways ngừng khai thác các chuyến bay. Có thể thấy vào năm các hãng hàng không ngừng khai thác chuyến bay thì số lượng hành khách giảm đột ngột hoặc tăng rất ít. 

## Câu hỏi 2: Tỉ lệ các chuyến bay đi đến các bang của Mỹ là bao nhiêu?

### 1. Mục đích
- Biết được, từ năm 2000-2020 các chuyến bay đến từng bang chiếm tỉ trọng bao nhiêu
- Từ đó biết được nên mở rộng thêm hay giảm bớt các chuyến bay đến bang nào

### 2. Phân tích

```python
dest_state_counts = df[df['DEST_STATE_NM'] != 'Unknown']['DEST_STATE_NM'].value_counts()

total_flights = dest_state_counts.sum()
dest_state_percentages = dest_state_counts / total_flights * 100

threshold = 2.2
small_states = dest_state_percentages[dest_state_percentages < threshold]
large_states = dest_state_percentages[dest_state_percentages >= threshold]

another_count = small_states.sum()
large_states['Another'] = another_count
```

### 3. Trực quan hóa
Sử dụng biểu đồ tròn, vì: Biểu đồ tròn cho thấy nhanh được tỷ lệ đóng góp

```python
plt.figure(figsize=(8, 6))
plt.pie(large_states, labels=large_states.index, labeldistance=1.05, autopct=lambda p: '{:.1f}%'.format(p) if p >= threshold else '', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Tỷ lệ các chuyến bay đi đến các bang của nước Mỹ')
plt.axis('equal')
plt.show()
```

### 4. Kết luận
- Các chuyến bay đến Alaska, California, Florida, Texas chiến phần lớn tỷ lệ, nguyên nhân có thể là do:
    - Các bang này có dân số đông, nên nhu cầu di chuyển về các bang này nhiều
    - Các bang có cơ sở hạ tầng tốt và sân bay lớn, ví dụ sân bay quốc tế Los Angeles ở California
    - Các bang có hoạt động kinh tế và thương mại sôi động, ví dụ Florida có nền kinh tế du lịch phát triển, và Alaska có ngành công nghiệp khai thác và đánh bắt cá quan trọng
- Như vậy, nhu cầu đi đến các bang kể trên lớn và có thể tập trung phát triển các chuyến bay đến các bang đó
