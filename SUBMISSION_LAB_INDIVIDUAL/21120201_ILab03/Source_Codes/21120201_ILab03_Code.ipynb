# Import necessary libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml

import warnings
warnings.filterwarnings('ignore')
```

# Load MNIST dataset

```python
# Load MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data
y = mnist.target.astype(int)
```

> Visualize MNIST dataset in 2D space using PCA and t-SNE, so we have `n_components = 2`.

# Perform t-SNE


Using **default values** in `sklearn` for important hyperparameters in t-SNE algorithm:
- The perplexity: `perplexity = 30.0`.
- The learning rate: `learning_rate = 'auto' = max(N / early_exaggeration / 4, 50)` with:
  - `N` is the sample size.
  - `early_exaggeration = 12.0` as default.
- The number of iterations: `n_iter = 1000` and `n_iter_without_progress = 300`.

```python
tsne = TSNE(n_components=2, random_state=0) # Using random_state=0 to have same solutions for multiple times of running code
X_tsne = tsne.fit_transform(X)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.cm.get_cmap('tab10', 10), s=10)
plt.colorbar(ticks=range(10))
plt.title('t-SNE Visualization of MNIST Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

# Perform PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=plt.cm.get_cmap('tab10', 10), s=10)
plt.colorbar(ticks=range(10))
plt.title('PCA Visualization of MNIST Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

> As we can see, t-SNE performs the clusters of data more effectively than PCA for MNIST dataset.
